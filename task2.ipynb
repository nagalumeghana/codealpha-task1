```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Create a sample dataset
data_dict = {
    "Date": ["2020-01-01", "2020-02-01", "2020-03-01", "2020-04-01", "2020-05-01"],
    "Unemployment_Rate": [4.0, 3.8, 4.2, 14.7, 13.3]
}
data = pd.DataFrame(data_dict)

# Save the dataset to a CSV file (for reference, not required for this script)
data.to_csv("unemployment_data.csv", index=False)

# Convert the 'Date' column to a datetime format
data['Date'] = pd.to_datetime(data['Date'])

# Display the first few rows
print("Dataset Preview:")
print(data.head())

# Basic statistics
print("\nDataset Statistics:")
print(data.describe())

# Visualize the trend of unemployment rate over time
plt.figure(figsize=(12, 6))
plt.plot(data['Date'], data['Unemployment_Rate'], marker='o', label='Unemployment Rate')
plt.title('Unemployment Rate Over Time')
plt.xlabel('Date')
plt.ylabel('Unemployment Rate (%)')
plt.legend()
plt.grid(True)
plt.show()

# Distribution of unemployment rate
plt.figure(figsize=(8, 6))
sns.histplot(data['Unemployment_Rate'], kde=True, bins=20, color='blue')
plt.title('Distribution of Unemployment Rate')
plt.xlabel('Unemployment Rate (%)')
plt.ylabel('Frequency')
plt.show()

# Example: Finding periods with the highest unemployment rate
highest_unemployment = data[data['Unemployment_Rate'] == data['Unemployment_Rate'].max()]
print("\nPeriods with the Highest Unemployment Rate:")
print(highest_unemployment)
```
